```python
# Importar a função data do pacote pydataset e outras bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import pydataset as data
```

```python
# Código do dataset
codigo_dataset = "Bundesliga"
```

```python
# Importando o dataset
df_liga = data.data(codigo_dataset)
```

```python
# Imprimindo na tela
df_liga.head()
```

```python
# 1. Verificar os jogos com maior quantidade de gols na história da Bundesliga

# Retorna as primeiras linhas do DataFrame df_liga e, em seguida, cria uma nova coluna 'Total_Gols'
# O valor de cada entrada nesta coluna é calculado somando os valores da coluna 'AwayGoals' e 'HomeGoals'

df_liga['Total_Gols'] = df_liga['AwayGoals'] + df_liga['HomeGoals']
```

```python
# Imprime as primeiras linhas do DataFrame modificado após a adição da nova coluna 'Total_Gols'
# O segundo head() limita a saída a apenas as primeiras linhas do DataFrame modificado

print(df_liga.head().head())
```

```python
# Classificando de ordem decrescente com base nos valores da coluna
df_liga.sort_values(by = 'Total_Gols', ascending = False)
```

```python
# Seleciona a primeira linha do DataFrame soccer_df após ordená-lo com base nos valores da coluna 'Total_Gols' em ordem decrescente.
# Isso significa que estamos obtendo a linha que representa o jogo com o maior número de gols.
df_liga.sort_values(by = 'Total_Gols', ascending = False).iloc[0]
```

```python
# Encontrar o máximo número de gols marcados em uma partida da liga
maximo = df_liga['Total_Gols'].max()

# Filtrar o DataFrame para obter as linhas onde o número total de gols é igual ao máximo
df_liga_filtered = df_liga[df_liga['Total_Gols'] == maximo]
```

```python
#Imprimindo a linha do dataframe
df_liga_filtered
```

```python
 #Criar uma cópia do DataFrame filtrado para evitar modificações indesejadas no original
 df_liga_filtered_copy = df_liga_filtered.copy()
```

```python
# Variável que armazena uma cópia do DataFrame filtrado para preservar o original
df_liga_filtered_copy
```

```python
# Converter a coluna 'Date' para o formato de data '%Y-%m-%d'  *coerce para tratar valores invalidos ex:  NaN
df_liga_filtered_copy['Date_New'] = pd.to_datetime(df_liga_filtered_copy['Date'] ,errors = 'coerce',format = '%Y-%m-%d').dt.strftime("%Y-%m-%d")

#Imprimindo a linha do dataframe
df_liga_filtered_copy
```
